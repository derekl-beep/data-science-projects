# Training & Test Dataset Split


```python
prices = data['PRICE']
features = data.drop('PRICE', axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, random_state=42)

# % of training set
len(X_train) / len(features)
```

```python
# % of test set
len(X_test) / len(features)
```

# Multivariate Regression

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
print('Intercept', regr.intercept_)
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'])
```

```python
# Challenge: print out the r-squred for training and test datasets
print('Training data r-squared: ', regr.score(X_train, y_train))
print('Test data r-squared: ', regr.score(X_test, y_test))
```

# Regression using log-prices


```python
prices = np.log(data['PRICE'])  # Use log-prices
features = data.drop('PRICE', axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, random_state=42)

regr = LinearRegression()
regr.fit(X_train, y_train)


print('Training data r-squared: ', regr.score(X_train, y_train))
print('Test data r-squared: ', regr.score(X_test, y_test))

print('Intercept', regr.intercept_)
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'])
```

```python
# Charles River Property Premium
np.e**0.096207
```

# p-values and Evaluating Coefficients

```python
X_incl_const = sm.add_constant(X_train)
X_incl_const.head()
```

```python
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

pd.DataFrame({'coef': results.params, 'p-values': round(results.pvalues, 3)})
```

# Testing for Multicolinearity


$$ TAX = \alpha_0 + \alpha_1 RM + \alpha_2 NOX + ... + \alpha_{12} LSTAT$$

$$ VIF_{TAX} = \frac{1}{(1 - R_{TAX}^2)} $$

```python
variance_inflation_factor(exog=X_incl_const.values, exog_idx=1)
```

```python
# Challenge: print all the VIFs for all the features

for i in range(X_incl_const.shape[1]):
    print(variance_inflation_factor(exog=X_incl_const.values, exog_idx=i))
```

```python
vif = [variance_inflation_factor(exog=X_incl_const.values, exog_idx=i) for i in range(X_incl_const.shape[1])]
pd.DataFrame({'coef_name': X_incl_const.columns, 'vif': np.around(vif, 2)})
```

# Model Simplification and Bayesian Information Criterion (BIC)

```python
# Original model with log-prices and all features

X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

org_coef = pd.DataFrame({'coef': results.params, 'p-values': round(results.pvalues, 3)})

# Challenge: print out the BIC and p-values
print('BIC: ' + str(results.bic))
print('R-squared: ' + str(results.rsquared))
```

```python
# Reduced model with log-prices, excluding INDUS

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'], axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

coef_minus_INDUS = pd.DataFrame({'coef': results.params, 'p-values': round(results.pvalues, 3)})

# Challenge: print out the BIC and p-values
print('BIC: ' + str(results.bic))
print('R-squared: ' + str(results.rsquared))
```

```python
# Reduced model with log-prices, excluding INDUS and AGE

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS', 'AGE'], axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

reduced_coef = pd.DataFrame({'coef': results.params, 'p-values': round(results.pvalues, 3)})

# Challenge: print out the BIC and p-values
print('BIC: ' + str(results.bic))
print('R-squared: ' + str(results.rsquared))
```

```python
frames = [org_coef, coef_minus_INDUS, reduced_coef]
pd.concat(frames, axis=1)
```
